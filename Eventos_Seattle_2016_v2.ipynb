```python
import urllib.request

def dl_data(url, output):
    try:        
        f = urllib.request.urlopen(url)
        print ("Downloading " + url)

        os.makedirs(os.path.dirname(output), exist_ok=True)
        with open(output, "wb") as local_file:
            local_file.write(f.read())

    except URLError:
        print ("Error", url)
```

### Según el día, queremos que revise si los ek tienen algún evento cercano

```python
# esta función es para calcular la distancia entre dis puntos pero no la EUCLIDEA sino teniendo en cuenta la curvatura de la
# tierra.
from math import radians, cos, sin, asin, sqrt, acos

def haversine(lon1, lat1, lon2, lat2):
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6373 # Radius of earth kilometers
    return c * r
```

```python
# voy a crear una lista de element_key con algun punto cercano (<=300 metros)
# una vez que un element_key tiene un punto cercano ya no miro si tiene otros (break)
df['event'] = 0
for df_index2, df_row2 in df.iterrows():
    df_lat2 = pd.to_numeric(df_row2['latitude'])
    df_lon2 = pd.to_numeric(df_row2['longitude'])
 #  df.at[df_index2,'event'] = 0
    for e_index, e_row in events.iterrows():
        if df_row2['timestamp'] == e_row['start_time']:
            e_lat1 = pd.to_numeric(e_row['Latitude'])
            e_lon1 = pd.to_numeric(e_row['Longitude'])
            dist = haversine(e_lon1, e_lat1, df_lon2, df_lat2)
            if dist <= 0.3:
                df.at[df_index2,'event'] = 1
                break
```
